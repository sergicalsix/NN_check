```python
import os
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,BatchNormalization,Conv2D,MaxPooling2D,Dropout
from tensorflow.keras.datasets import mnist,cifar10
from tensorflow.keras.utils import to_categorical

epoch = 100

(x_train,y_train),(x_test,y_test) = cifar10.load_data()
x_train = np.float32(x_train.reshape(50000,32,32,3))/255.0
y_train = to_categorical(y_train,10)
x_test = np.float32(x_test.reshape(10000,32,32,3))/255.0
y_test = to_categorical(y_test,10)



model = Sequential([
    Conv2D(5,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu',kernel_initializer='he_normal',input_shape=(32\
,32,3)),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(10,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu',kernel_initializer='he_normal'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(400,activation='relu',kernel_initializer='he_normal'),
    Dropout(0.5),
    Dense(400,activation='relu',kernel_initializer='he_normal'),
    Dropout(0.5),
    Dense(10,activation='softmax',kernel_initializer='he_normal')
])
print(model.summary())

#model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
#history = model.fit(x_train,y_train,batch_size=100,epochs=epoch,verbose=0,validation_data=(x_test,y_test))
#print(history.history)
```

```python
model = Sequential([
    Conv2D(20,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu',kernel_initializer='he_normal',input_shape=(32\
,32,3)),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(50,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu',kernel_initializer='he_normal'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(400,activation='relu',kernel_initializer='he_normal'),
    Dropout(0.5),
    Dense(400,activation='relu',kernel_initializer='he_normal'),
    Dropout(0.5),
    Dense(10,activation='softmax',kernel_initializer='he_normal')
])
print(model.summary())
```
